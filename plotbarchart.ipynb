```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 数据
labels = ['Original Simulation Result', 'Actual Boiler Unit Value', 'Adjusted Boiler Unit Value']
values = [12.21, 8.42, 6.19]
actual_value = values[1]

# 设置字体
plt.rcParams['font.family'] = 'Times New Roman'

# 绘制条形图
plt.figure(figsize=(12, 8))
bars = plt.bar(labels, values, color=['#1f77b4', '#ff7f0e', '#2ca02c'], edgecolor='black', linewidth=1.2)

# 设置图表标题和标签
plt.title('Energy Consumption Comparison (2015/11/1~11/18)', fontsize=18, fontweight='bold')
plt.ylabel('MWh', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 添加网格线
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 显示数值标签和百分比差异
for bar in bars:
    yval = bar.get_height()
    label = f'{yval:.2f} MWh'
    
    # 计算与实际值的百分比差异
    if bar.get_label() != 'Actual Boiler Unit Value':
        difference_percentage = ((yval - actual_value) / actual_value) * 100
        label += f'\n({difference_percentage:.2f}%)'
    
    # 调整数值标签的位置
    plt.text(bar.get_x() + bar.get_width() / 2, yval - 0.3, label, ha='center', va='bottom', fontsize=12)

# 调整图表的布局
plt.tight_layout()

# 显示图表
plt.show()
```
